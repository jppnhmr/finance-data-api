# Interest Rates and Commodity Prices

Exploring how U.S. interest rates and global commodity prices have moved together over time using simple rolling correlations.

## What this notebook is about

I built this notebook to practice working with real economic time-series data and to explore how relationships between variables can change over time.

Instead of looking at a single correlation across the full dataset, I use rolling correlations to see how the relationship between U.S. interest rates and commodity prices behaves during different periods such as:
- the 2008 financial crisis
- the COVID shock
- the post-pandemic inflation period

This is exploratory analysis, not a predictive model.

## Data used

- U.S. policy interest rate (monthly)
- Global commodity price indices:
  - Energy
  - Food
  - All commodities

All series are aligned to monthly frequency so they can be compared directly.

## Approach

For each commodity index, I calculate a 36-month rolling Pearson correlation with U.S. interest rates.

Using a rolling window helps show how the relationship changes over time, instead of assuming it is constant across different economic conditions.

## Interpreting the patterns (high level)

The charts above show that the relationship between interest rates and commodity prices changes noticeably across different periods. Rather than treating this as a single relationship, it is more useful to think in terms of broad economic regimes.

**20082009: Financial Crisis**  
During the global financial crisis, commodity prices and interest rates move in opposite directions. This lines up with a period of collapsing demand and rapid policy easing. In this environment, interest rates appear to be reacting to economic conditions rather than driving commodity prices.

**20102012: Early post-crisis recovery**  
In the years following the crisis, the relationship temporarily becomes more positive. This period overlaps with large stimulus efforts outside the U.S., including Chinas infrastructure and credit expansion, which supported global commodity demand even while U.S. interest rates remained very low. This provides a possible explanation for the brief strengthening in correlation.

**20122017: Low-rate, uneven growth period**  
As interest rates remained near zero for an extended time, correlations weaken and become less stable. This suggests that when policy rates are constrained, they may provide limited signal, and commodity prices are more influenced by supply conditions and global demand factors.

**20172019: Expansion and gradual tightening**  
During the late-2010s expansion, interest rates and commodity prices tend to move more closely together. This is consistent with a period of steady growth, where rising demand and gradual policy tightening occur at the same time.

**2020: COVID-19 shock**  
The pandemic introduces sharp and short-lived changes in the relationship. Lockdowns, sudden demand drops, and supply disruptions all occur at once, making simple correlations harder to interpret. Food prices in particular appear less responsive during this period, likely due to inelastic demand and supply constraints.

**20222024: Post-pandemic inflation period**  
In the most recent period, correlations turn strongly negative, especially for energy and broad commodities. This coincides with rising interest rates alongside slowing commodity momentum. One possible interpretation is that tighter financial conditions and weaker demand begin to matter more, although supply and geopolitical factors are also important.

Overall, these patterns suggest that interest rates and commodity prices do not have a fixed relationship. The broader economic environment plays a large role in shaping how these variables move together over time.
